```python
# Importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Carregando os dados
df = pd.read_csv('data/processed/data.csv')  # Certifique-se de que o caminho do arquivo está correto

# Exibindo as primeiras linhas do DataFrame para verificar os dados
print(df.head())
```

```python
# Estatísticas descritivas
print(df.describe())

# Verificando valores nulos
print(df.isnull().sum())
```

```python
# Plotando a distribuição do Fluxo de Caixa Líquido
plt.figure(figsize=(10,6))
sns.histplot(df['Fluxo de Caixa Líquido (R$)'], kde=True, bins=20)
plt.title('Distribuição do Fluxo de Caixa Líquido')
plt.xlabel('Fluxo de Caixa Líquido (R$)')
plt.ylabel('Frequência')
plt.show()

# Removendo colunas não numéricas para plotar a matriz de correlação
numerical_df = df.select_dtypes(include=[np.number])

# Plotando correlações entre variáveis numéricas
plt.figure(figsize=(12,8))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlação entre Variáveis')
plt.show()
```
